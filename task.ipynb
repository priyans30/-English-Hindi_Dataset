```python
from datasets import load_dataset
import pandas as pd

ds = load_dataset("ainlpml/english-hindi")
df = pd.DataFrame(ds['train'])
print(df.shape)
print(df.head())
```

```python
import re

def is_hindi(text):
    devanagari_chars = re.findall(r'[\u0900-\u097F]', str(text))
    return len(devanagari_chars) / max(1, len(text)) > 0.2

df['is_hindi'] = df['text'].apply(is_hindi)
```

```python
df_en = df[df['is_hindi'] == False].reset_index(drop=True)
df_hi = df[df['is_hindi'] == True].reset_index(drop=True)
```

```python
pairs = pd.DataFrame({
    'english': df_en['text'],
    'hindi': df_hi['text']
})
```

```python
pairs
```

```python
pairs = pairs.dropna(subset=['english', 'hindi'])
```

```python
pairs
```

```python
x = 900
print(pairs["english"][x])
print(pairs["hindi"][x])
```

```python
# Function to count words
def word_count(text):
    return len(str(text).split())

# Add word counts
pairs['word_count_en'] = pairs['english'].apply(word_count)
pairs['word_count_hi'] = pairs['hindi'].apply(word_count)
```

```python
pairs
```

```python
pairs = pairs[
    (pairs['word_count_en'].between(5, 50)) &
    (pairs['word_count_hi'].between(5, 50))
].reset_index(drop=True)
```

```python
pairs['diff'] = pairs['word_count_en'] - pairs['word_count_hi']

pairs = pairs[pairs['diff'].between(-10, 10)].reset_index(drop=True)
```

```python
pairs
```

```python
x = 78
print(pairs["english"][x])
print(pairs["hindi"][x])
```

```python
pairs.to_excel("final_cleaned_dataset.xlsx", index=False)
```

```python
english_sentences = pairs['english'].head(100).tolist()
```

```python
import getpass
import os

if "GROQ_API_KEY" not in os.environ:
    os.environ["GROQ_API_KEY"] = getpass.getpass("enter_api_key")
```

```python
from langchain_groq import ChatGroq

# Initialize ChatGroq LLM
llm = ChatGroq(
    model="openai/gpt-oss-20b",
)

```

```python
hindi_translations = []

for sent in english_sentences:
    response = llm.invoke(
       (f"Translate this English sentence to Hindi:\n\n{sent}")
    )
    hindi_translations.append(response.content)  # get the translated text

# Optional: convert to DataFrame for convenience
import pandas as pd
translation_df = pd.DataFrame({
    'english': english_sentences,
    'hindi_pred': hindi_translations
})
```

```python
translation_df.to_excel("translation_output.xlsx", index=False)
```

```python
import sacrebleu

# Reference translations
references = [pairs['hindi'].head(100).tolist()]  # list of list

# BLEU
bleu = sacrebleu.corpus_bleu(hindi_translations, references)
print("BLEU:", bleu.score)

# CHRF
chrf = sacrebleu.corpus_chrf(hindi_translations, references)
print("CHRF:", chrf.score)

# TER
ter = sacrebleu.corpus_ter(hindi_translations, references)
print("TER:", ter.score)
```
