```python
from datasets import load_dataset
import pandas as pd

ds = load_dataset("ainlpml/english-hindi")
df = pd.DataFrame(ds['train'])
print(df.shape)
print(df.head())
```

```python
import re

def is_hindi(text):
    # if more than 30% of chars are Devanagari (U+0900–U+097F)
    devanagari_chars = re.findall(r'[\u0900-\u097F]', str(text))
    return len(devanagari_chars) / max(1, len(text)) > 0.2

df['is_hindi'] = df['text'].apply(is_hindi)
```

```python
df_en = df[df['is_hindi'] == False].reset_index(drop=True)
df_hi = df[df['is_hindi'] == True].reset_index(drop=True)
```

```python
pairs = pd.DataFrame({
    'english': df_en['text'],
    'hindi': df_hi['text']
})
```

```python
import re
import pandas as pd
from datasets import load_dataset


ds = load_dataset("ainlpml/english-hindi")
df = pd.DataFrame(ds['train'])


def devanagari_ratio(text):

    text = str(text)
    devanagari_chars = re.findall(r'[\u0900-\u097F]', text)
    return len(devanagari_chars) / max(1, len(text))

df['devanagari_ratio'] = df['text'].apply(devanagari_ratio)


df = df.sort_values(by='devanagari_ratio', ascending=True).reset_index(drop=True)

df_en = df.head(10000).copy()
df_hi = df.tail(10000).copy()

```

```python
df_hi = df_hi.iloc[::-1]
```

```python
df_hi
```

```python
import pandas as pd

# Pair English and Hindi rows
paired = pd.DataFrame({
    'english': df_en['text'].reset_index(drop=True),
    'hindi': df_hi['text'].reset_index(drop=True)
})
```

```python
# Function to count words
def word_count(text):
    return len(str(text).split())

# Add word counts
paired['word_count_en'] = paired['english'].apply(word_count)
paired['word_count_hi'] = paired['hindi'].apply(word_count)
```

```python
paired
```

```python
paired.shape
```

```python
paired = paired[
    (paired['word_count_en'].between(5, 50)) &
    (paired['word_count_hi'].between(5, 50))
].reset_index(drop=True)
```

```python
paired['diff'] = paired['word_count_en'] - paired['word_count_hi']

paired = paired[paired['diff'].between(-10, 10)].reset_index(drop=True)
```

```python
paired.to_excel("final_cleaned_dataset.xlsx", index=False)
```
